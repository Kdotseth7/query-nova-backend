### Add User

```python
import requests
import json
from pprint import pprint

data = {
    "name": "joker3",
    "email": "joker3@email.com"
}

headers = {'Content-Type': 'application/json'}

r = requests.post('http://127.0.0.1:8000/users', data=json.dumps(data), headers=headers)

pprint(r.text, indent=2)
```

### Query

```python
import requests
import json
from pprint import pprint

data = {
    "user_id": "1",
    "query": "Provide all usr table records"
}

headers = {'Content-Type': 'application/json'}

r = requests.post('http://127.0.0.1:8000/query', data=json.dumps(data), headers=headers)

pprint(r.text, indent=2)
```
